#**Scaling**


#With Normalization

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
```

```python
DF = sns.load_dataset('mpg')
```

```python
DF.info()
```

```python
DF.head()
```

```python
X = DF[['weight']]
y = DF['mpg']
```

```python
X[:5]
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize= (9,6))
sns.regplot(x=X,y=y)
plt.xlabel('weight_Without_Scaling')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler1 = MinMaxScaler()
X_Norm = scaler1.fit_transform(X)

```

```python
X_Norm[:5]
```

```python
fig = plt.figure(figsize = (9,6))
sns.regplot(x= X_Norm, y=y)
plt.xlabel('weight_with_normalization')
plt.show()
```

#With Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
X_stan = scaler2.fit_transform(X)

```

```python
X_stan[:5]
```

```python
fig= plt.figure(figsize = (9,6))
sns.regplot(x=X_stan, y=y)
plt.xlabel('weight_with_standardization')
plt.show()
```
